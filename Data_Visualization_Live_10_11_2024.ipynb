```python
cat_cols = list(data.dtypes[data.dtypes=='object'].index)
num_cols =list(data.dtypes[data.dtypes!='object'].index)
```

Impute missing values

for numerical cols-->Midian/Mean <br>
for categorical cols -->Mode

```python
for cols in num_cols:
  med= data[cols].median()
  data[cols] = data[cols].fillna(med)
```

```python
for cols in cat_cols:
  mode = data[cols].mode()[0]
  data[cols] = data[cols].fillna(mode)
```

```python
data[cat_cols].isna().sum()
```

```python
data[num_cols].isna().sum()
```

## Outliers Treatment



*   Drop the outliers
*   replacing with median
*   Capping and Flooring(box plot method)




```python
col = 'Lotfrontage'
```

```python
 for col in num_cols:
  q1 =data[col].describe()["25%"]
  q3 = data[col].describe()["75%"]
  iqr = q3-q1
  lower_whisker = q1-1.5*iqr
  upper_whisker = q3+1.5*iqr
  data[col] = np.where(data[col]>upper_whisker,upper_whisker,data[col])
  data[col] = np.where(data[col]<lower_whisker,lower_whisker,data[col])
```

```python
if data[col].max()>upper_whisker or data[col].min()<lower_whisker:
  print("Outliers are present")
else:
  print("Outliers are not present")
```
